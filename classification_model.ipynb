<h1>Classificação de Plantas da Espécie Iris com ScikitLearn<h1\>

```python
from sklearn.datasets import load_iris
```

Importando dataset iris, dados de classificação de plantas da espécie iris

```python
iris = load_iris()
```

```python
iris.data
```

```python
iris.target
```

```python
iris.target_names
```

```python
list(iris.target_names)
```

Atribuindo valores ás variáveis X (dados) e y (target = classificação)

```python
X,y = load_iris(return_X_y=True)
```

```python
X
```

```python
y
```

Transformando o dataset em um dataframe

```python
import pandas as pd
```

```python
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
```

```python
iris_df
```

Adicionando a coluna com as classificações

```python
iris_df['target'] = iris.target
```

```python
iris_df.head()
```

```python
iris_df['target names'] = pd.Categorical.from_codes(iris.target, iris.target_names)
```

```python
iris_df
```

Plotando o gráfico da classificação em função do tamanho das sépalas e pétalas

```python
%matplotlib inline
```

```python
iris_df.plot.scatter('sepal length (cm)', 'sepal width (cm)', c='target')
```

```python
iris_df.plot.scatter('petal length (cm)', 'petal width (cm)', c='target')
```

Divisão do Dateset em treinamento e teste:

```python
from sklearn.model_selection import train_test_split
```

Modelo recebe 33% de linhas de teste e 66% de linhas de treinamento

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

100 linhas de treinamento

```python
X_train.shape
```

50 linhas de teste

```python
X_test.shape
```

Classificação com SVM:

Plotar em pares as características do dataframe

```python
import seaborn as sns
```

```python
sns.pairplot(iris_df[['sepal length (cm)', 'sepal width (cm)',
                     'petal length (cm)','petal width (cm)',
                     'target names']], hue='target names')
```

É possível notar intuitivamente que são linearmente separávies

Utilizando classificador SVM(Máquinas de Vetores de Suporte)

```python
from sklearn import svm
```

Criar um classificador e definir a margem rígida ou suave

```python
clf = svm.SVC(C=1.0)
```

Treinamento do modelo com os dados de treinamento

```python
clf.fit(X_train,y_train)
```

Predição do conjunto de teste

```python
clf.predict(X_test)
```

Medindo os resultados

```python
y_test
```

```python
clf.score(X_test,y_test)
```

Métricas de avaliação de um classificador:

```python
from sklearn.metrics import classification_report
```

```python
y_pred = clf.predict(X_test)
```

Medidas de precisã, revocação, pontuação quantidade de amostras e acurácia

```python
print(classification_report(y_test, y_pred, target_names=iris.target_names))
```
